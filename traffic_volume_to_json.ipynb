```python
# traffic volume
traffic_filename = "data/traffic_volume/TrafficVol_ 05-Present2014.csv"

# street network
folder = "data/street_network_database/WGS84/"
shp = "Street Network Database.shp"
streets_filename = folder + shp
#path = "output/"
path = "../gmap/data/"

out_filename = 'streetnet_traffic_volumes_bounded'

if False:
    xlim1 = -122.29
    xlim2 = -122.27
    ylim1 = 47.53 #47.6
    ylim2 = 47.55 #47.63
if False:
    xlim1 = -122.304
    xlim2 = -122.258
    ylim1 = 47.51 #47.6
    ylim2 = 47.58 #47.63
    
if True:
    xlim1=float("-inf")
    ylim1=float("-inf")
    xlim2=float("inf")
    ylim2=float("inf")
```

```python
import fiona
import json
from shapely.geometry import LineString, Polygon

figure(figsize=(14,14))

features = []
crs = None
with fiona.collection(streets_filename, "r") as source:
    count = 0
    for feat in source:
        
        compkey = feat['properties']['COMPKEY']
        
        if compkey not in volumes.keys():
            continue
        
        #feat['properties'].update(...) # with your attributes
        pt1 = feat['geometry']['coordinates'][0]

        if not (pt1[0] > xlim1 and pt1[0] < xlim2 and \
                pt1[1] > ylim1 and pt1[1] < ylim2):
            continue
        
        sorted_keys = sorted(volumes[compkey].keys(), reverse=True)
        awdt = volumes[compkey][sorted_keys[0]]['awdt']
        
        #print feat['geometry']
        coords = [(pt[0], pt[1]) for pt in feat['geometry']['coordinates']]
        line = LineString(coords)
        sc = 0.00007
        poly = line.buffer(sc + awdt/2500.0 * sc, cap_style=2)
        #print poly.exterior
        if poly.type == 'MultiPolygon':
            continue
        x,y = poly.exterior.xy
        plot(x,y)
        #plot([pt[0] for pt in coords], [pt[1] for pt in coords])
        #z = [0 for pt in x]
        
        #plot(line.xy)
        #print line.xy
        #print feat['geometry']['coordinates']
        feat['geometry']['type'] = 'Polygon'
        new_coords = [zip(x, y)]
        feat['geometry']['coordinates'] = new_coords
        #print feat.keys()
        #print feat['type']
        if False:
            print feat.keys() #len(feat['geometry']['coordinates'][0])
            print (new_coords[0])
            print ' '
        
        #first_key = sorted(volumes[compkey].keys(), reverse=True)[0]
        #print 'key ', first_key
        artclass = feat['properties']['ARTCLASS']
        feat['properties'] = {}
        feat['properties']['ARTCLASS'] = artclass
        
        count = 1000
        for key in sorted_keys:
            prefix = 'v' + str(count)[1:] + '_'
            val = "{0:.2f}".format(key) + ' ' + "{0:.0f}".format(volumes[compkey][key]['awdt']) + ' ' + volumes[compkey][key]['dirflow']
            feat['properties'][prefix + 'date'] = val
            #feat['properties'][prefix + 'awdt'] = volumes[compkey][key]['awdt']
            #feat['properties'][prefix + 'dirflow'] = volumes[compkey][key]['dirflow']
            count += 1
            
        #volumes[compkey]
        
        features.append(feat)
        
        #print volumes[compkey]
        
        count += 1
        #if count > 5:
        #    break
    crs = " ".join("+%s=%s" % (k,v) for k,v in source.crs.items())

street_layer = {
    "type": "FeatureCollection",
    "features": features,
    "crs": {
        "type": "link", 
        "properties": {"href": out_filename + ".crs", "type": "proj4"} }}

if True:
    with open(path + out_filename + ".json", "w") as f:
        f.write(json.dumps(street_layer, sort_keys=True))
    with open(path + out_filename + ".crs", "w") as f:
        f.write(crs)
        
```
