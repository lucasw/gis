```python
# traffic volume
filename = "data/traffic_volume/TrafficVol_ 05-Present2014.csv"

# street network
folder = "data/street_network_database/WGS84/"
shp = "Street Network Database.shp"
streets_filename = folder + shp

if True:
    out_filename = 'streetnet_traffic_volumes_bounded'
    xlim1 = -122.29
    xlim2 = -122.27
    ylim1 = 47.53 #47.6
    ylim2 = 47.55 #47.63
if False:
    out_filename = 'streetnet_traffic_volumes_bounded'
    xlim1 = -122.304
    xlim2 = -122.258
    ylim1 = 47.51 #47.6
    ylim2 = 47.58 #47.63
    
if False:
    out_filename = 'streetnet_traffic_volumes'
    xlim1=float("-inf")
    ylim1=float("-inf")
    xlim2=float("inf")
    ylim2=float("inf")
```

```python
import fiona
import json
from shapely.geometry import LineString, Polygon

figure(figsize=(14,14))

features = []
crs = None
with fiona.collection(streets_filename, "r") as source:
    count = 0
    for feat in source:
        #feat['properties'].update(...) # with your attributes
        pt1 = feat['geometry']['coordinates'][0]

        if not (pt1[0] > xlim1 and pt1[0] < xlim2 and \
                pt1[1] > ylim1 and pt1[1] < ylim2):
            continue
            
        #print feat['geometry']
        coords = [(pt[0], pt[1]) for pt in feat['geometry']['coordinates']]
        line = LineString(coords)
        poly = line.buffer(0.00015, cap_style=2)
        #print poly.exterior
        x,y = poly.exterior.xy
        plot(x,y)
        plot([pt[0] for pt in coords], [pt[1] for pt in coords])
        #z = [0 for pt in x]
        
        #plot(line.xy)
        #print line.xy
        #print feat['geometry']['coordinates']
        feat['geometry']['type'] = 'Polygon'
        feat['geometry']['coordinates'] = [zip(x, y)]
        #print feat.keys()
        #print feat['type']
        #print feat
        features.append(feat)
        
        count += 1
        #if count > 2:
        #    break
    crs = " ".join("+%s=%s" % (k,v) for k,v in source.crs.items())

street_layer = {
    "type": "FeatureCollection",
    "features": features,
    "crs": {
        "type": "link", 
        "properties": {"href": "street_net.crs", "type": "proj4"} }}

if True:
    with open("output/" + out_filename + ".json", "w") as f:
        f.write(json.dumps(street_layer))
    with open("output/" + out_filename + ".crs", "w") as f:
        f.write(crs)
        
```
