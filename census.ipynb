```python
def make_ogr_point(x,y):
    return ogr.Geometry(wkt="POINT(%f %f)"%(x,y))

# http://stackoverflow.com/questions/21335091/python-area-of-irregular-polygon-results-in-negative-value
# assume last point and first are same, apparently common in survey/gis data
def get_area(pts):
    area = 0.0

    n = len(pts)
    # if matching endpoint isn't true, maybe should make -1 optional
    for i in range(n - 1):
        i1 = (i+1)%n
        area += pts[i][0]*pts[i1][1] - pts[i1][0]*pts[i][1]       
        area *= 0.5

    return abs(area);
```

```python
# http://www.arachnoid.com/area_irregular_polygon/index.html
def find_area(array):
    a = 0
    ox,oy = array[0]
    for x,y in array[1:]:
        a += (x*oy-y*ox)
        ox,oy = x,y
    return a/2
```

```python
def get_pts(geom):
    x1=float("inf")
    y1=float("inf")
    x2=float("-inf")
    y2=float("-inf")

    ch = geom.ConvexHull()
    if not ch:
        return None
    bd = ch.GetBoundary()
    if not bd:
        return None
    pts = bd.GetPoints()
    if not pts:
        return None

    for x,y in pts:
        x1 = min(x1,x)
        y1 = min(y1,y)
        x2 = max(x2,x)
        y2 = max(y2,y)
        
    return (pts, (x1, y1, x2, y2))
```

```python
# open the file, convert gdal to pts, limit geographical area
def process(input_filename, output_filename, lims):

    print "limits ", lims
    xlim1 = lims[0]
    ylim1 = lims[1]
    xlim2 = lims[2]
    ylim2 = lims[3]
    
    # open the shapefile
    ds = ogr.Open( input_filename )
    if ds is None:
        print "Open failed.\n"
        sys.exit( 1 )

    lyr = ds.GetLayerByIndex( 0 )

    lyr.ResetReading()

    feat_defn = lyr.GetLayerDefn()
    field_defns = [feat_defn.GetFieldDefn(i) for i in range(feat_defn.GetFieldCount())]

    # look up the index of the field we're interested in
    for i, defn in enumerate( field_defns ):
        if defn.GetName()=="POP10":
            pop_field = i

    n_features = len(lyr)
  
    # TODO get shp files of King County, Seattle, other interesting boundaries:
    # zip codes, legislative districts, precincts, Seattle districts...
    # and test if points are in interior or not, provide boundary file from
    # command line

    x1 = float("inf")
    y1 = float("inf")
    x2 = float("-inf")
    y2 = float("-inf")
    
    census_pre = []
    
    # mainly find the bounds, clip to lims
    for j, feat in enumerate( lyr ):

        pop = feat.GetField(pop_field)
        geom = feat.GetGeometryRef()
        if geom is None:
            continue
        
        dat = get_pts(geom)
        if dat is None:
            continue
        pts, bb = dat
        #if len(pts) != 4:
        #    print j, " ", pop, " ", len(pts)

        #exclude region outside of limits
        if (bb[0] > xlim2) or (bb[0] < xlim1) or (bb[1] > ylim2) or (bb[1] < ylim1):
            continue

        x1 = min(x1, bb[0])
        y1 = min(y1, bb[1])
        x2 = max(x2, bb[2])
        y2 = max(y2, bb[3])
        
        census_pre.append((bb, pts, pop))
    
    bbg = x1, y1, x2, y2
    print "bounding box ", bbg
    
        
    # TODO split here
    return census_pre, bbg
```

```python
def proc_pop_and_flim(census_pre, bbg, flims):    
    
    census = []
    x1b = float("inf")
    y1b = float("inf")
    x2b = float("-inf")
    y2b = float("-inf")
  
    print "fractional limits ", flims
    xflim1 = flims[0]
    yflim1 = flims[1]
    xflim2 = flims[2]
    yflim2 = flims[3]
    
    density_min = float("inf")
    density_max = float("-inf")
    #density_avg = 
    area_tot = 0

    pop_min = float("inf")
    pop_max = float("-inf")
    pop_tot = 0
    
    x1 = bbg[0]
    y1 = bbg[1]
    x2 = bbg[2]
    y2 = bbg[3]
    dx = x2 - x1
    dy = y2 - y1
    
    for block in census_pre:
        bb = block[0]
        pts = block[1]
        pop = block[2]
        
        x1f = (bb[0] - x1) / dx
        y1f = (bb[1] - y1) / dy
        #x2f = (bb[2] - x1) / dx
        #y2f = (bb[3] - y1) / dy
        if (x1f > xflim2) or (x1f < xflim1) or \
           (y1f > yflim2) or (y1f < yflim1):
            continue
        
        x1b = min(x1b, bb[0])
        y1b = min(y1b, bb[1])
        x2b = max(x2b, bb[2])
        y2b = max(y2b, bb[3])
      
        #exclude region outside of limits
       
        # all the pts end with with the same starting point, need to remember this
        # when computing area
        #if (j == 0):
        #    print "points ", pts
    
        # TBD need a lat/long to meters conversion
        area = find_area(pts)
        
        if (area <= 0):
            print j, " bad area ", pts
            continue
        
        area_tot += area

        #if (area < 2e-10):
        #    print "small area", j, pop, area, pts
       
        density = 0
        if (pop > 0):
            density = pop/area
            density_min = min(density_min, density)
            density_max = max(density_max, density)
        
        census.append((bb, pts, pop, area, density))
        
        pop_min = min(pop_min, pop)
        pop_max = max(pop_max, pop)
        pop_tot += pop

        #if j%1000==0:
        #    #conn.commit()
        #    print "%s/%s (%0.2f%%)"%(j+1,n_features,100*((j+1)/float(n_features)))
        #    print "population ", pop
        #    print bb
   
    

    bbb = x1b, y1b, x2b, y2b
    print "clipped bounding box ", bbb
    print "pop ", pop_tot, ", bounds ", pop_min, pop_max
    print "density ", density_min, density_max, ", area total ", area_tot 
    
    census_sorted = sorted(census, key=lambda block: block[4]) 
    
    print "num fractionally clipped blocks", len(census_sorted),\
        "out of", j, "geographically clipped"
    
    return census_sorted, bbb, pop_tot, density_max, area_tot
```

```python
def process2(census_sorted, bb, pop_tot, density_max, area_tot):
    
    x1 = bb[0]
    y1 = bb[1]
    x2 = bb[2]
    y2 = bb[3]

    dy = y2 - y1
    dx = (x2 - x1) * math.cos(y1 * math.pi/180.0)
    ratio = dy / dx
    print "ratio ", ratio
    width = 4096
    height = ratio * width

    pix_per_deg = width / dx 
    print "pix per deg ", pix_per_deg
     
    # when this reach pop_tot/2, shift coloring
    pop_accum = 0
    # TODO later make number of subdivision dynamic
    pop_area = [0, 0]
    density_div = -1

    #print len(census_sorted)
    #print census_sorted[0]
    #print census_sorted[0][0]
      
    pix_pts = []
    
    for block in census_sorted:
        pop = block[2]
        pop_accum += pop
        area = block[3]
        density = block[4]

        deg_pts = block[1]
        pts = []
        for (xd,yd) in deg_pts:
            xp = (xd - x1) * pix_per_deg * math.cos(y1 * math.pi / 180.0) 
            yp = (yd - y1) * pix_per_deg
            pts.append((xp, yp)) 
        
        if pop > 0:
          # change colorization based on which half of the population density
          # divide the block is in
          if pop_accum > pop_tot / 2:
            # keep track of boundary
            if density_div < 0:
                density_div = density         
            pop_area[0] += area
          else:
            pop_area[1] += area
        
        pix_pts.append( pts )
   
    print "area ", [x / area_tot for x in pop_area]
    
    #screen.blit(surf, (0,0))
    #pygame.image.save(screen, sys.argv[2])
    #time.sleep(0.0)
    
    return pix_pts, density_div
```
