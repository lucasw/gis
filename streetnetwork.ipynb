```python
def get_pts(geom):
    x1=float("inf")
    y1=float("inf")
    x2=float("-inf")
    y2=float("-inf")

    #ch = geom.ConvexHull()
    #if not ch:
    #    return None
    #bd = ch.GetBoundary()
    #if not bd:
    #    return None
    pts = geom.GetPoints()
    if not pts:
        return None

    #print len(pts)
    #print pts
    # z is probably supposed to be elevation, but is -1.7976931348623157e+308
    for x,y,z in pts:
        x1 = min(x1,x)
        y1 = min(y1,y)
        x2 = max(x2,x)
        y2 = max(y2,y)

    return (pts, (x1, y1, x2, y2))
```

```python
# open the file, convert gdal to pts, limit geographical area
def get_pt_sets(input_filename, output_filename, lims):

    print "limits ", lims
    xlim1 = lims[0]
    ylim1 = lims[1]
    xlim2 = lims[2]
    ylim2 = lims[3]
    
    # open the shapefile
    ds = ogr.Open( input_filename )
    if ds is None:
        print "Open failed.\n", input_filename
        sys.exit( 1 )

    lyr = ds.GetLayerByIndex( 0 )

    lyr.ResetReading()

    feat_defn = lyr.GetLayerDefn()
    field_defns = [feat_defn.GetFieldDefn(i) for i in range(feat_defn.GetFieldCount())]

    # look up the index of the field we're interested in
    
    f_inter_ind = 0
    t_inter_ind = 0
    for i, defn in enumerate( field_defns ):
        if defn.GetName()=="F_INTR_ID":
            f_inter_ind = i
        if defn.GetName()=="T_INTR_ID":
            t_inter_ind = i

    n_features = len(lyr)
    print "num features", n_features
  
    # TODO get shp files of King County, Seattle, other interesting boundaries:
    # zip codes, legislative districts, precincts, Seattle districts...
    # and test if points are in interior or not, provide boundary file from
    # command line

    length_min = float("inf")
    length_max = float("-inf")
    length_tot = 0
    
    x1 = float("inf")
    y1 = float("inf")
    x2 = float("-inf")
    y2 = float("-inf")
    
    census_pre = []
    
    # mainly find the bounds, clip to lims
    for j, feat in enumerate( lyr ):

        #if (j > 10):
        #    break
        
        inter_1 = feat.GetField(f_inter_ind)
        inter_2 = feat.GetField(t_inter_ind)
            
        geom = feat.GetGeometryRef()
        if geom is None:
            print  j, " no geom"
            continue
        
        dat = get_pts(geom)
        if dat is None:
            print j, "no pts"
            continue
        pts, bb = dat
        
        #print j, inter_1, inter_2, len(pts), geom.Length()
        length = geom.Length()
        length_min = min(length, length_min)
        length_max = max(length, length_max)
        length_tot+= length
        
        #exclude region outside of limits
        if (bb[0] > xlim2) or (bb[0] < xlim1) or (bb[1] > ylim2) or (bb[1] < ylim1):
            continue

        x1 = min(x1, bb[0])
        y1 = min(y1, bb[1])
        x2 = max(x2, bb[2])
        y2 = max(y2, bb[3])
        
        census_pre.append((bb, pts, length))
    
    bbg = x1, y1, x2, y2
    print "bounding box ", bbg
    print "length min max", length_min, length_max, length_tot
    
        
    # TODO split here
    return census_pre, bbg, length_min, length_max
```

```python
def proc_pop_and_flim(census_pre, bbg, flims):    
    
    census = []
    x1b = float("inf")
    y1b = float("inf")
    x2b = float("-inf")
    y2b = float("-inf")
  
    print "fractional limits ", flims
    if (flims != None):
        xflim1 = flims[0]
        yflim1 = flims[1]
        xflim2 = flims[2]
        yflim2 = flims[3]
    
    x1 = bbg[0]
    y1 = bbg[1]
    x2 = bbg[2]
    y2 = bbg[3]
    dx = x2 - x1
    dy = y2 - y1
    
    for block in census_pre:
        bb = block[0]
        pts = block[1]
        length = block[2]
        
        x1f = (bb[0] - x1) / dx
        y1f = (bb[1] - y1) / dy
        x2f = (bb[2] - x1) / dx
        y2f = (bb[3] - y1) / dy
        # outside of bounds
        if (flims != None):
            if (x1f > xflim2) and (x2f > xflim2) or \
               (x1f < xflim1) and (x2f < xflim1) or \
               (y1f > yflim2) and (y2f > yflim2) or \
               (y1f < yflim1) and (y2f < yflim1):
                continue
        
        x1b = min(x1b, bb[0])
        y1b = min(y1b, bb[1])
        x2b = max(x2b, bb[2])
        y2b = max(y2b, bb[3])
 
    
        census.append((bb, pts, length))

    bbb = x1b, y1b, x2b, y2b
    print "clipped bounding box ", bbb
    
    census_sorted = census #sorted(census, key=lambda block: block[4]) 
    
    print "num fractionally clipped blocks", len(census_sorted),\
        "out of", len(census_pre), "geographically clipped"
    
    return census_sorted, bbb
```

```python
def process2(census_sorted, bb, length_tot):
    
    x1 = bb[0]
    y1 = bb[1]
    x2 = bb[2]
    y2 = bb[3]

    dy = y2 - y1
    dx = (x2 - x1) * math.cos(y1 * math.pi/180.0)
    ratio = dy / dx
    print "ratio ", ratio
    width = 4096
    height = ratio * width

    pix_per_deg = width / dx 
    print "pix per deg ", pix_per_deg
      
    pix_pts = []
    
    for block in census_sorted:
        length = block[2]
        
        deg_pts = block[1]
        pts = []
        for (xd,yd,zd) in deg_pts:
            xp = (xd - x1) * pix_per_deg * math.cos(y1 * math.pi / 180.0) 
            yp = (yd - y1) * pix_per_deg
            pts.append((xp, yp)) 
        
        pix_pts.append( pts )
    
    return pix_pts
```
