```python
def propagate_graph_all(graph, streets, iterations):
    #if iterations < 0:
    
    # initialize all the intersections to weight/rank/score 1.0
    for j, key in enumerate(graph.keys()):
        intr = graph[key]      
        intr['weight'] = 1.0
        intr['weight_new'] = 0.0
    
    for i in range(iterations):
        graph, min_weight, max_weight = propagate_graph(graph, streets)
    
    print "graph propagation", min_weight, max_weight
    
    return graph, min_weight, max_weight

# function to call repeatedly
def propagate_graph(graph, streets):
    length_min = 0.0003
    length_max = length_min * 15.0   
    
    for j, intr_id in enumerate(graph.keys()):
        intr = graph[intr_id]
        
        for snd_id in intr['snd_id']:
  
            if not streets.has_key(snd_id):
                print "error could not find snd_id", snd_id
                continue
                
            (feat, inter_1, inter_2) = streets[snd_id]
            
            if (inter_1 != intr_id and inter_2 != intr_id):
                print "bad intersection", intr_id, inter_1, inter_2, snd_id
                continue
                
            intr_id_far = inter_1
            if inter_1 == intr_id:
                intr_id_far = inter_2
                
            if not graph.has_key(intr_id_far):
                print "bad intersection id", snd_id, intr_id_far
                continue
            intr_far = graph[intr_id_far]
            
            geom = feat.GetGeometryRef()
            if geom is None:
                print  j, key, compkey, " no geom"
                continue
                
            length = geom.Length()
            # clip the length on low end
            length = max(length, length_min)
            # rescale 
            length /= length_min
            
            weight_fraction = 0.5 ** length
            
            intr_far['weight_new'] += weight_fraction * intr_far['weight']
            
    # go through all the intersections and add weight_new to weight, 
    # set weight_new to zero
    min_weight = float("inf")
    max_weight = float("-inf")
    
    for j, key in enumerate(graph.keys()):
        intr = graph[key]      
        intr['weight'] += intr['weight_new']
        intr['weight_new'] = 0.0
        
        min_weight = min(min_weight, intr['weight'])
        max_weight = max(max_weight, intr['weight'])
        
    
    return graph, min_weight, max_weight
```

```python
def proc_pop_and_flim(census_pre, bbg, flims):    
    
    census = []
    x1b = float("inf")
    y1b = float("inf")
    x2b = float("-inf")
    y2b = float("-inf")
  
    print "fractional limits ", flims
    if (flims != None):
        xflim1 = flims[0]
        yflim1 = flims[1]
        xflim2 = flims[2]
        yflim2 = flims[3]
    
    x1 = bbg[0]
    y1 = bbg[1]
    x2 = bbg[2]
    y2 = bbg[3]
    dx = x2 - x1
    dy = y2 - y1
    
    for block in census_pre:
        bb = block[0]
        pts = block[1]
        length = block[2]
        
        x1f = (bb[0] - x1) / dx
        y1f = (bb[1] - y1) / dy
        x2f = (bb[2] - x1) / dx
        y2f = (bb[3] - y1) / dy
        # outside of bounds
        if (flims != None):
            if (x1f > xflim2) and (x2f > xflim2) or \
               (x1f < xflim1) and (x2f < xflim1) or \
               (y1f > yflim2) and (y2f > yflim2) or \
               (y1f < yflim1) and (y2f < yflim1):
                continue
        
        x1b = min(x1b, bb[0])
        y1b = min(y1b, bb[1])
        x2b = max(x2b, bb[2])
        y2b = max(y2b, bb[3])
 
    
        census.append((bb, pts, length))

    bbb = x1b, y1b, x2b, y2b
    print "clipped bounding box ", bbb
    
    census_sorted = sorted(census, key=lambda block: block[2]) 
    
    print "num fractionally clipped blocks", len(census_sorted),\
        "out of", len(census_pre), "geographically clipped"
    
    return census_sorted, bbb
```

```python
def process2(census_sorted, bb, length_tot):
    
    x1 = bb[0]
    y1 = bb[1]
    x2 = bb[2]
    y2 = bb[3]

    dy = y2 - y1
    dx = (x2 - x1) * math.cos(y1 * math.pi/180.0)
    ratio = dy / dx
    print "ratio ", ratio
    width = 4096
    height = ratio * width

    pix_per_deg = width / dx 
    print "pix per deg ", pix_per_deg
      
    pix_pts = []
    
    for block in census_sorted:
        length = block[2]
        
        deg_pts = block[1]
        pts = []
        for (xd,yd,zd) in deg_pts:
            xp = (xd - x1) * pix_per_deg * math.cos(y1 * math.pi / 180.0) 
            yp = (yd - y1) * pix_per_deg
            pts.append((xp, yp)) 
        
        pix_pts.append( pts )
    
    return pix_pts
```

```python
def adj(val, th, ind):
    return (val - th[ind])/(th[ind+1] - th[ind])
```

```python
def conv_ll(latlong, fr):
    longitude = latlong[0]
    latitude = latlong[1]
    if (fr == None):
        fr = math.cos(latitude * math.pi / 180.0)
        print "latitude scale fraction", fr
    x = longitude * fr
    y = latitude
    
    return (x, y, fr)
```

```python
# pass in a 0-1.0 value
def color_scale(val):
    
    th = [0, 0.05, 0.1, 0.5, 0.8, 1.0]
    
    alpha = 0.5
    col = (0,0,0, alpha)
    if (val > th[4]):
        v2 = adj(val, th, 4)
        col = (0, v2, 0.5 - val/2.0, alpha)
    elif (val > th[3]):
        v2 = adj(val, th, 3)
        col = (0, 1.0 - val, val, alpha)
    elif (val > th[2]):
        v2 = adj(val, th, 2)
        col = (0, 0.5 + val/2.0, 0, alpha)
    elif (val > th[1]):
        v2 = adj(val, th, 2)
        col = (1.0 - val, val/2.0, 0, alpha)
    else:
        v2 = adj(val, th, 0)
        col = (val, 0, 0, alpha)
        
    return col
```

```python
# draw all the streets connected to this intersection
def draw_streets(graph, intr, fr):
    for snd_id in intr['snd_id']:
        
        (feat, inter_1, inter_2) = streets[snd_id]
        latlong1 = graph[inter_1]['latlong']
        x1, y1, fr = conv_ll(latlong1, fr)
        latlong2 = graph[inter_2]['latlong']
        x2, y2, fr = conv_ll(latlong2, fr)
        plot([x1,x2], [y1,y2], linewidth = 0.3, color = (0,0,0,0.1))
```
